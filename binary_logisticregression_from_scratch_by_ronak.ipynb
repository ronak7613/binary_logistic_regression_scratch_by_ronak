```python
import numpy as np

class LogisticRegression:
    def __init__(self, lr=0.01, num_iter=100000, fit_intercept=True, regularization='none', reg_strength=0.01, batch_size=32):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.regularization = regularization
        self.reg_strength = reg_strength
        self.batch_size = batch_size
    
    def add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    def l1_regularization(self):
        return self.reg_strength * np.sum(np.abs(self.theta[1:]))
    
    def l2_regularization(self):
        return (self.reg_strength / 2) * np.sum(self.theta[1:] ** 2)
    
    def no_regularization(self):
        return 0
    
    def get_regularization_function(self):
        if self.regularization == 'l1':
            return self.l1_regularization
        elif self.regularization == 'l2':
            return self.l2_regularization
        elif self.regularization == 'none':
            return self.no_regularization
        else:
            raise ValueError('Invalid regularization type.')
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = self.add_intercept(X)
        
        self.theta = np.zeros((X.shape[1],1))
        regularization_function = self.get_regularization_function()
        
        for i in range(self.num_iter):
            batch_indices = np.random.randint(X.shape[0], size=self.batch_size)
            X_batch = X[batch_indices]
            y_batch = y[batch_indices]
            
            z = np.dot(X_batch, self.theta)
            h = self.sigmoid(z)
            gradient = np.dot(X_batch.T, (h - y_batch)) / y_batch.size + regularization_function()
            gradient[0] -= regularization_function() # no regularization on bias term
            #print(gradient.shape,'and',self.lr)
            self.theta -= self.lr * gradient
            
            
    
    def predict_prob(self, X):
        if self.fit_intercept:
            X = self.add_intercept(X)
    
        return self.sigmoid(np.dot(X, self.theta))
    
    def predict(self, X, threshold=0.5):
        return self.predict_prob(X) >= threshold
```
